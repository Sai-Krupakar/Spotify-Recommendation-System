```python
import csv
import numpy as np
import statistics
FV1 =np.array([])
FV2 =np.array([])
FV3 =np.array([])
FV4 =np.array([])
FV5 =np.array([])
FV6 =np.array([])
FV7 =np.array([])
FV8 =np.array([])
FV9 =np.array([])
FV10 =np.array([])
FV11 =np.array([])
FV12 =np.array([])
# opening the CSV file
with open('data.csv', mode ='r')as file:
   
  # reading the CSV file
  csvFile = csv.reader(file)
 
  # displaying the contents of the CSV file
  count = 0
  for lines in csvFile:
    if count == 0:
      count = count+1
      continue
    FV1=np.append(FV1,float(lines[0]))
    FV2 = np.append(FV2,float(lines[2]))
    FV3 = np.append(FV3,float(lines[4]))
    FV4 = np.append(FV4,int(lines[5]))
    FV5 = np.append(FV5,float(lines[6]))
    FV6 = np.append(FV6,float(lines[9]))
    FV7 = np.append(FV7,float(lines[11]))
    FV8 = np.append(FV8,float(lines[12]))
    FV9 = np.append(FV9,float(lines[15]))
    FV10 = np.append(FV10,float(lines[17]))
    FV11 = np.append(FV11,float(lines[18]))
print(FV1)
FV1 = (FV1-statistics.mean(FV1))/statistics.stdev(FV1)
print(FV1)
FV2 = (FV2-statistics.mean(FV2))/statistics.stdev(FV2)
FV3 = (FV3-statistics.mean(FV3))/statistics.stdev(FV3)
FV4 = (FV4-statistics.mean(FV4))/statistics.stdev(FV4)
FV5 = (FV5-statistics.mean(FV5))/statistics.stdev(FV5)
FV6 = (FV6-statistics.mean(FV6))/statistics.stdev(FV6)
FV7 = (FV7-statistics.mean(FV7))/statistics.stdev(FV7)
FV8 = (FV8-statistics.mean(FV8))/statistics.stdev(FV8)
FV9 = (FV9-statistics.mean(FV9))/statistics.stdev(FV9)
FV10 = (FV10-statistics.mean(FV10))/statistics.stdev(FV10)
FV11 = (FV11-statistics.mean(FV11))/statistics.stdev(FV11)
print(FV11)
```

```python
FV1 = np.reshape(FV1,(170653,1))
FV2 = np.reshape(FV2,(170653,1))
FV3 = np.reshape(FV3,(170653,1))
FV4 = np.reshape(FV4,(170653,1))
FV5 = np.reshape(FV5,(170653,1))
FV6 = np.reshape(FV6,(170653,1))
FV7 = np.reshape(FV7,(170653,1))
FV8 = np.reshape(FV8,(170653,1))
FV9 = np.reshape(FV9,(170653,1))
FV10 = np.reshape(FV10,(170653,1))
FV11 = np.reshape(FV11,(170653,1))
print(np.shape(FV1))
FV = np.concatenate((FV1,FV2,FV3,FV4,FV5,FV6,FV7,FV8,FV9,FV10,FV11),axis = 1)
print(np.shape(FV))
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
sse = {}
for k in range(1,10):
 kmeans = KMeans(n_clusters=k)
 kmeans.fit(FV)
 sse[k] = kmeans.inertia_
plt.plot(list(sse.keys()), list(sse.values()))
plt.show()
plt.scatter(FV1,FV2,c=kmeans.labels_,s = 0.1)
plt.show()
print(kmeans.labels_[1])
```
